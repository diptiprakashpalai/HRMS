```python
# Create DataFrame with the rows which has ID information that is every alternative rows
ID_RAWS= df[df[1] == 'ID:']
# Only select the columns in the DataFrame where ID value is oresent that is third column
ID = ID_RAWS.iloc[:, 2]
# Reset the Index of newly made DataFrame
ID1 = ID.reset_index(drop = True)
print (ID1)
```

```python
# Here it is shown how to split the In & Out Hour present in a shell of the excel sheet
# One particular shell is selected
hr = df_with_ID.iloc[30,1]
print (hr)
# Convert the string to List
hr1 = list(hr)
# First 5 elements are selected
hr_t1 = hr1[:5]
# Last 5 elements are selected
hr_t2 = hr1[-5:]
# Join the elements in each List to form the string
hr2 = ' '.join(hr_t1)
hr3 = ' '.join(hr_t2)
print (hr2)
print (hr3)
print (type(hr2))
```

```python
# Remove the spaces from the strings
hr4 = hr2.replace(' ', '')
hr5 = hr3.replace(' ', '')
# Convert String to Time
import datetime
hr2_time = datetime.datetime.strptime(hr4, "%H:%M").time()
hr3_time = datetime.datetime.strptime(hr5, "%H:%M").time()
date = datetime.date(1, 1, 1)
print (hr2_time)
print (hr3_time)
# Difference of In & Out Time
print (datetime.datetime.combine(date, hr3_time) - datetime.datetime.combine(date, hr2_time))
```
